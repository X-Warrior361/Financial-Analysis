```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
```

```python
data_2022 = pd.read_csv("equity_data_2022.csv")
data_2020 = pd.read_csv("equity_data_2020.csv")
```

```python
data_2020.columns = data_2020.columns.str.lower()
data_2022.columns = data_2022.columns.str.lower()
data_2020 = data_2020.drop(['series'], axis=1)
data_2020 = data_2020.rename(columns={'date1': 'date'})
data_2022 = data_2022.rename(columns={'date1': 'date'})

data_2020.date = pd.to_datetime(data_2020.date)
data_2022.date = pd.to_datetime(data_2022.date)
```

### Working on Reliance

```python
reliance = data_2020[data_2020.symbol=="RELIANCE"]
reliance.head()
```

```python
reliance = reliance.sort_values('date')
reliance
```

```python
reliance.date.value_counts()
```

```python
reliance = reliance.drop_duplicates()
reliance.date.value_counts()
```

```python
#Plotting the time series data
plt.figure(figsize=(15,10))
plt.plot(reliance.date, reliance.close_price ,label='Close Price')
#plt.plot(reliance.date, reliance.open_price ,label='Open Price')
plt.title('Reliability of Reliance Industries Limited')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```
